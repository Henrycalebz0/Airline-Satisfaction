```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/Lenovo/Desktop/docs/Airline Satisfaction/airline_passenger_satisfaction.csv')
```

```python
#set index

df.set_index('ID')
```

```python
#fill nul values with default placeholder values

df['Arrival Delay'] = df['Arrival Delay'].fillna(0)
```

```python
df.info()
```

```python
#summary of the dataset

data_description = df.describe()

satisfaction_count = df['Satisfaction'].value_counts()

data_description
```

```python
#satisfcation percentage

percentage = df['Satisfaction'].value_counts(normalize=True) * 100
round(percentage, 2)
```

```python
#Count of customer types
customer_grouped = df.groupby('Customer Type').agg({
    'Customer Type':'count'
})
customer_grouped
```

```python
#Count of type of travel
type_grouped = df.groupby('Type of Travel').agg({
    'Type of Travel':'count'
})

type_grouped
```

```python
#Count of type of travel
class_grouped = df.groupby('Class').agg({
    'Class':'count'
})
class_grouped
```

```python
#correlation

df['Satisfaction_binary'] = df['Satisfaction'].apply(lambda x: 1 if x == 'Satisfied' else 0)

numeric_df = df.select_dtypes(include=[np.number])

correlation_matrix = numeric_df.corr()

correlation_matrix
```

```python
#satisfacttion correlation

satisfaction_corr = correlation_matrix['Satisfaction_binary'].drop('Satisfaction_binary').sort_values(ascending=False)
satisfaction_corr
```
